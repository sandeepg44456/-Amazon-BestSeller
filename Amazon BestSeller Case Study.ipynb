**5)What is the average price of the books?**



```python
#Plotting a boxplot for checking for any outliers in price attribute
sns.boxplot(df,x='Price')
plt.show()
```

```python
#Creating a function for removing outliers
def outliers(column):
    percentile25=column.quantile(0.25)
    percentile75=column.quantile(0.75)
    iqr=percentile75-percentile25
    lower_range=percentile25-1.5*iqr
    upper_range=percentile75+1.5*iqr
    return lower_range,upper_range
```

```python
outliers(df['Price'])
```

```python
pd.reset_option('display.max_rows',None)
removed_price_outlier=df[(df['Price']>-6.5) & (df['Price']<29.5)]
removed_price_outlier
```

```python
#Plotting a boxplot to check for outliers
sns.boxplot(removed_price_outlier,x='Price')
plt.show()
```

**15)How has the number of bestseller books published per year changed over time?**

```python
df1=df[['Name','Year','User Rating']]
df1=df1.drop_duplicates()
df2=df1.groupby('Year')['User Rating'].mean().reset_index()
df2
```

```python
#Plotting a line graph
plt.figure(figsize=(10,6))
sns.lineplot(df2,x='Year',y='User Rating',linewidth=4,color='magenta')
sns.scatterplot(df2,x='Year',y='User Rating',s=150,color='magenta')
plt.title("Average User Rating of BestSellers Over the Years",fontstyle='italic',fontsize=14)
plt.ylabel("Average User Rating")
plt.show()
```

```python
df1=df[['Name','Year','Reviews']]
df1=df1.drop_duplicates()
df2=df1.groupby('Year')['Reviews'].sum()
df2=df2.reset_index()
df2
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(df2,x='Year',y='Reviews',color='cyan',linewidth=4)
sns.scatterplot(df2,x='Year',y='Reviews',color='cyan',s=150)
plt.title("Total Number of Reviews Each Year",fontsize=14,fontstyle='italic')
plt.ylabel("Total Reviews")
plt.show()
```

* The average rating and the total number of reviews have **increased** over the years for the bestsellers.
